# Data Cleaning

```python
# Dropping null values

movies=data.dropna()
movies
```

```python
#Dropping columns

movie=movies.drop(['Unnamed: 0','Director'],axis=1)
movie
```

# Removing Special Characters

```python
movie['Genre'] = movie['Genre'].apply(lambda x:x.replace('\n',''))
```

```python
movie['Movie Name'] = movie['Movie Name'].apply(lambda x:x.replace('\n',''))
```

```python
movie['Year'] = movie['Year'].apply(lambda x:x.replace('\n',''))
```

```python
movie['Votes'] = movie['Votes'].apply(lambda x:x.replace(',',''))
```

```python
movie['Year'] = movie['Year'].apply(lambda x:x.replace(')',''))
```

```python
movie['Gross in M $'] = movie['Gross'].apply(lambda x:x.replace('$',''))
```

```python
movie['Gross in M $'] = movie['Gross in M $'].apply(lambda x:x.replace('M',' '))
```

```python
movie
```

# Removing Numbers from Movie name column

```python
movie['Movie Name'][0]
```

```python
movie['Movie Name'][0].replace('1.','')
```

```python
p=re.compile(r"\.(.*)")
```

```python
p.findall(movie['Movie Name'][0])
```

```python
m=movie['Movie Name'].apply(lambda x:p.findall(x))
```

```python
m=m.apply(lambda x:"".join(x))
```

```python
movie['Movie Name']=m
```

```python
movie
```

```python
movie.drop(['Gross'],axis=1,inplace=True)
movie
```

# Removing duplicate values

```python
duplicates=movie.drop_duplicates()
duplicates
```

```python
movie['Run Time'] = movie['Run Time'].apply(lambda x:x.replace(' min',''))
```

```python
movie
```

```python
movie.dtypes
```

```python
movie['Rating']=movie['Rating'].astype('float')
movie['Votes']=movie['Votes'].astype('int')
movie['Run Time']=movie['Run Time'].astype('int')
movie['Gross in M $']=movie['Gross in M $'].astype('float')
```

```python
movie['Year'].value_counts()
```

```python
movie['Year'] = movie['Year'].apply(lambda x:x.replace('  ',''))
```

```python
movie['Year'] = movie['Year'].apply(lambda x:x.replace(' '' ',''))
```
