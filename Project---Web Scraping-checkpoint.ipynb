# Web Scraping on IMDB

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import requests
from bs4 import BeautifulSoup

import re
```

```python
title = []
year = []
runtime = []
certificates =[]
rating = []
genre = []
directors = []
Movie_year = []
for i in range(1,7):

    url = 'https://www.imdb.com/list/ls006266261/?sort=list_order,asc&st_dt=&mode=detail&page={}'.format(i)

    page = requests.get(url)
    pagecontent = page.text
    soup=BeautifulSoup(pagecontent)
    list =soup.find_all('div',attrs={'class':'lister-item mode-detail'})

    for j in list:
        
        #Name
        name = j.find('h3',attrs={'class':'lister-item-header'})
        if name is None:
            title.append(np.NaN)
        else:
            title.append(name.text.split(('('))[0])

        #Length
        length = j.find('span',attrs={'class':'runtime'})
        if length is None:
            runtime.append(np.NaN)
        else:
            runtime.append(length.text)
        
        #Year
        years = j.find('h3', attrs={'class' : 'lister-item-header'})
        if years is None:
            year.append(np.NaN)
        else:
            year.append(years.text.split(('('))[1])

        #Certificate
            certificate = j.find('span',attrs={'class':'certificate'})
        if certificate is None:
            certificates.append(np.NaN)
        else:
            certificates.append( certificate.text)

        #Ratings
        ratings = j.find('span',attrs={'class':'ipl-rating-star__rating'})
        if ratings is None:
            rating.append(np.NaN)
        else:
            rating.append(ratings.text)

        #Geners
        genres = j.find('span',attrs={'class':'genre'})
        if genres is None:
            genre.append(np.NaN)
        else:
            genre.append(genres.text)

        #Director
        director=j.find('p',attrs={'class':'text-muted text-small'})

        if director is None:
            directors.append(np.NaN)
        else:
            directors.append(director.text)
```

```python
votes = []
#gross = []


for i in range(1,7):
    url ='https://www.imdb.com/list/ls050782187/={}'. format(i)
    page = requests.get(url)
    htmlCode = page.text
    for j in soup.find_all('div',attrs = {"class":"lister-item-content"}):
        k =j.find_all("span",attrs={"name":"nv"})
        if k is None:
            votes.append(np.NaN)
        else:
            votes.append(k)
```

```python
v=[]
for i in votes:
    if i[0] is None:
        v.append(np.NaN)
    else:
        v.append(i[0].text)
```
